```python
import pandas as pd
```

```python
df = pd.read_csv('data1.csv')
```

```python
# Преобразуем 'GPSTime' в формат datetime (если это строка)
df['GPS Time'] = pd.to_datetime(df['GPS Time'])

# Вычитаем 7 секунд
df['GPSTime_Adjusted'] = df['GPS Time'] - pd.Timedelta(seconds=7)

df.head()
```

```python
# Смещаем GPSTime_Adjusted на 3 строки вверх
df['GPSTime_Adjusted'] = df['GPSTime_Adjusted'].shift(3)

# Выводим для проверки
print(df[[' Device Time', 'GPSTime_Adjusted']].head(20))
```

```python
df = df.iloc[3:]
print(df[[' Device Time', 'GPSTime_Adjusted']].head(20))
```

```python
# Убедимся, что время в секундах
df['GPSTime_Adjusted'] = pd.to_datetime(df['GPSTime_Adjusted'])  # Преобразуем в datetime
```

```python
print(df[[' Device Time', 'GPSTime_Adjusted']].head(20))
```

```python
# Определяем, какие строки содержат первое уникальное значение скорости GPS в каждой секунде
df['is_first'] = df.duplicated(subset=['GPSTime_Adjusted'], keep='first')  # Первое значение = False, остальные = True
print(df[[' Device Time', 'GPSTime_Adjusted', 'is_first']].head(20))
```

```python
# Заменяем повторяющиеся значения скорости GPS на nan
df.loc[df['is_first'], 'Speed (GPS)(km/h)'] = pd.NA
# Убираем вспомогательные столбцы
df.drop(columns=['GPS Time', 'is_first', 'GPSTime_Adjusted'], inplace=True)
```

```python
df['Speed (GPS)(km/h)'] = pd.to_numeric(df['Speed (GPS)(km/h)'], errors='coerce')
df['Speed (OBD)(km/h)'] = pd.to_numeric(df['Speed (OBD)(km/h)'], errors='coerce')
```
